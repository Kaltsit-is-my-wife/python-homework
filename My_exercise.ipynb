```
import random

n = int(input("please enter the number of random numbers: "))
randList = []
for i in range(n):
    r = random.randint(1,999)
    randList.append(r)

randList.sort()

print(randList)
```

```
n = int(input("please enter the number of numbers: "))
intList = []
for i in range(n):
    r = int(input("please enter the number: "))
    intList.append(r)

print("the average is: %.3f" % (sum(intList)/len(intList)))
```
